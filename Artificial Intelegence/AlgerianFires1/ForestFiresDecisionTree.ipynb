```python
import pandas as pd
```

```python
data = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', sep='\\s*,\\s*', skipinitialspace=True, skiprows=[0,124,125,126], engine='python')
```

```python
data.head()
```

```python
data['Classes'] = data['Classes'].replace(['not fire', 'fire'], [0, 1])

numerical_columns = ["Temperature", "RH", "Ws", "Rain", "FFMC", "DMC", "DC", "ISI", "BUI", "FWI"]
for column in numerical_columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')


input_variables = ["Temperature", "RH", "Ws", "Rain"]
fwi_variables = ["FFMC", "DMC", "DC", "ISI", "BUI"]
fwi = ["FWI"]

data = data.dropna(subset=input_variables + fwi_variables + fwi + ['Classes'])
```

```python
from sklearn.model_selection import train_test_split

X1 = data[input_variables]
X2 = data[fwi_variables]
X3 = data[fwi]
X4 = data[input_variables + fwi_variables + fwi]

Y = data['Classes']

X1_train, X1_test, Y_train, Y_test = train_test_split(X1, Y, test_size=0.2, random_state = 42)
X2_train, X2_test, Y_train, Y_test = train_test_split(X2, Y, test_size=0.2, random_state = 42)
X3_train, X3_test, Y_train, Y_test = train_test_split(X3, Y, test_size=0.2, random_state = 42)
X4_train, X4_test, Y_train, Y_test = train_test_split(X4, Y, test_size=0.2, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

clf = DecisionTreeClassifier()
clf2 = clf.fit(X2_train, Y_train)
```

```python
Y2_pred = clf2.predict(X2_test)

accuracy2 = accuracy_score(Y_test, Y2_pred)
print(f"Accuracy: {accuracy2:.2f}")

print(classification_report(Y_test, Y2_pred))

# Print the confusion matrix
print(confusion_matrix(Y_test, Y2_pred))
```
